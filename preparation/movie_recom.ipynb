## Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df3=df1.copy()
```

```python
df3['genres'] = df3['genres'].str.replace(r"[\[\]']", "", regex=True)
df3['keywords'] = df3['keywords'].str.replace(r"[\[\]']", "", regex=True)
```

```python
df3.loc[0]
```

```python
def one_text(row):
  text = []
  if pd.notna(row['genres']):
    text.append(row['genres'])

  if pd.notna(row['keywords']):
    text.append(row['keywords'])

  if pd.notna(row['overview']):
   text.append(row['overview'])


  return " ".join(text)

```

```python
df3['text'] = df3.apply(one_text, axis=1)
```

```python
df3.loc[1]
```

```python
movie_texts = df3['text'].to_list()
```

```python
movie_texts[0]
```

```python
movie_ids = df3['id'].to_list()
movie_ids[0]
```

```python
movie_titles = df3['title'].to_list()
movie_titles[0]
```

TF-IDF

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = vectorizer.fit_transform(movie_texts)
```

```python
print(tfidf_matrix.shape)
```

```python
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
similarity_matrix
```

```python
model_data = {
    'movie_ids': movie_ids,
    'movie_titles': movie_titles,
    'similarity_matrix': similarity_matrix.tolist()
}

with open('model_similarity.json', 'w') as f:
    json.dump(model_data, f)
```

```python
def find_similar_movies(target_movie_id, top_n=6):
    try:
        movie_index = movie_ids.index(target_movie_id)
    except ValueError:

        return []

    similarities = similarity_matrix[movie_index]

    similar_indices = np.argsort(similarities)[-top_n-1:-1][::-1]

    similar_movies = []
    for idx in similar_indices:
        similar_movies.append({
            'id': movie_ids[idx],
            'title': movie_titles[idx],
            'similarity_score': float(similarities[idx])
        })

    return similar_movies
```

```python
similar = find_similar_movies(1084242, top_n=5)

for movie in similar:
    print(f" {movie['title']}, {movie['id']}, {movie['similarity_score']:.3f}")
```
